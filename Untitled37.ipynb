```python
#Data Cleaning and Exploratory Data Analysis (EDA) Code
#importing Liabraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV file
file_path = 'titanic.csv'
data = pd.read_csv(file_path)
```

```python
print(data)
```

```python
# Data Cleaning: Handling Missing Values and Correcting Data Types

# Fill missing values for 'Age' with the median age
data['Age'] = data['Age'].fillna(data['Age'].median())
```

```python
# Fill missing value for 'Fare' with the median fare
data['Fare'] = data['Fare'].fillna(data['Fare'].median())
```

```python
# For 'Cabin', fill missing values with a placeholder 'Unknown'
data['Cabin'] = data['Cabin'].fillna('Unknown')
```

```python
# Verify that all missing values are handled
missing_values_after_cleaning = data.isnull().sum()
print("Missing Values After Cleaning:")
print(missing_values_after_cleaning)
```

```python
# Exploratory Data Analysis (EDA)

# Generate summary statistics for numerical columns
numerical_summary = data.describe()
print("\nNumerical Summary:")
print(numerical_summary)
```

```python
# Generate summary statistics for categorical columns
categorical_summary = data.describe(include=['object'])
print("\nCategorical Summary:")
print(categorical_summary)
```

```python
# More Pandas Operations

# Count the number of survivors and non-survivors
survival_counts = data['Survived'].value_counts()
print("\nSurvival Counts:")
print(survival_counts)
```

```python
# Survival rate
survival_rate = data['Survived'].mean()
print("\nSurvival Rate:")
print(survival_rate)
```

```python
# Average age of survivors vs non-survivors
average_age_survivors = data[data['Survived'] == 1]['Age'].mean()
average_age_non_survivors = data[data['Survived'] == 0]['Age'].mean()
print("\nAverage Age of Survivors:")
print(average_age_survivors)
print("Average Age of Non-Survivors:")
print(average_age_non_survivors)
```

```python
#Survival rate by sex
survival_rate_by_sex = data.groupby('Sex')['Survived'].mean()
print("\nSurvival Rate by Sex:")
print(survival_rate_by_sex)
```

```python
# Survival rate by passenger class
survival_rate_by_pclass = data.groupby('Pclass')['Survived'].mean()
print("\nSurvival Rate by Passenger Class:")
print(survival_rate_by_pclass)
```

```python
# Visualizations

# Plotting distributions for numerical features
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], kde=True, bins=30)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Fare'], kde=True, bins=30)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting survival rate by sex
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Survived', data=data)
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Plotting survival rate by passenger class
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', data=data)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Plotting survival rate by embarkation port
plt.figure(figsize=(10, 6))
sns.barplot(x='Embarked', y='Survived', data=data)
plt.title('Survival Rate by Embarkation Port')
plt.xlabel('Port of Embarkation')
plt.ylabel('Survival Rate')
plt.show()
```
